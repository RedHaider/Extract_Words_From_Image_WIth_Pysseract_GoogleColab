```python
from google.colab import drive
import pytesseract
import shutil
import os
import random
from PIL import Image
from docx import Document
```

```python
from docx import Document
import re

def clean_text(text):
    # Remove control characters and non-XML compatible characters
    cleaned_text = re.sub(r'[\x00-\x08\x0B-\x0C\x0E-\x1F\x7F-\x9F]', '', text)
    return cleaned_text

image_path_in_colab = '/content/Sample-Job-Application-Letter.jpg'

# Extract text from the image
extractedInformation = pytesseract.image_to_string(Image.open(image_path_in_colab))

# Clean the extracted text
cleaned_text = clean_text(extractedInformation)

# Create a new document
document = Document()

# Add the cleaned text to the document
document.add_paragraph(cleaned_text)

output_path = "/content/Sample-Job-Application-Letter.docx"

# Save the document
document.save(output_path)

print("Document saved successfully.")
```
