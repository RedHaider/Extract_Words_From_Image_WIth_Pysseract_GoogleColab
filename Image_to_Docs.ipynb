```python
import pytesseract
from PIL import Image
from docx import Document
from google.colab import drive
drive.mount('/content/drive')

def clean_text(text):
    # Remove non-XML compatible characters and control characters
    cleaned_text = "".join(ch if ord(ch) < 128 and ch.isprintable() else " " for ch in text)
    return cleaned_text

def image_to_doc(image_path, output_path):
    image = Image.open(image_path)
    extracted_text = pytesseract.image_to_string(image)

    # Clean the extracted text
    cleaned_text = clean_text(extracted_text)

    document = Document()
    document.add_paragraph(cleaned_text)
    document.save(output_path)

image_path = '/content/drive/MyDrive/ADN_DATA/Image_to_docs/Sample-Job-Application-Letter.jpg'
output_path = '/content/drive/MyDrive/ADN_DATA/Docs_to_Images/Sample-Job-Application-Letter.docx'

image_to_doc(image_path, output_path)
```
